```python
import numpy as np
import pandas as pd

# Removing duplicates

df_data = pd.DataFrame({"k1": ["one", "two"] * 3  + ["two"], "k2": [1, 1, 2, 3, 3, 4, 4]})
df_data
```

```python
# CHECKS FOR DUPLICATES

print(df_data.duplicated())
```

```python
# REMOVE THE DUPLICATES

df_data.drop_duplicates()
```

```python
# ADD A COLUMN AND DROP THE DUPLICATES BY SUBSET

df_data["v1"] = range(7)
df_data
```

```python
df_data.duplicated()
```

```python
df_data.duplicated(subset=["k1"])
```

```python
#REMOVE THE DUPLICATES BASED ON THE TWO SUBSETS

df_data.drop_duplicates(subset=["k1", "k2"])
```

```python
df_food = pd.DataFrame({"food" : ["bacon", "pork", "pastram", "tomato", "mushroom", "hamburger", "corn beef hash", "pulled pork"],
                           "ounces" : [4, 3,5, 10, 6, 7.5, 8, 9]})

df_food
```

```python
# ADD A COLUMN, WHICH IS A DICTIONARY ENTRY

df_food["tastiness"] = [10, 7, 3, 5, 4, 5, 6, 9]

df_food
```

```python
#MAPPING A DICTIONARY

meat_to_animal = {"bacon": "pig", "pork": "pig", "pastram": "pig", "tomato": "veggie", "mushroom": "veggie", "hamburger": "beef", "corn beef hash": "beef", "pulled pork": "pork"}

df_food["animal"]  = df_food["food"].map(meat_to_animal)

df_food
```

```python
# MAPPING A DICTIONARY USING A FUNCTION

df_food_revised = pd.DataFrame({"food" : ["bacon", "pork", "pastram", "tomato", "mushroom", "hamburger", "corn beef hash", "pulled pork"],
                           "ounces" : [4, 3,5, 10, 6, 7.5, 8, 9]})

meat_to_animal_revised = {"bacon": "pig", "pork": "pig", "pastram": "pig", "tomato": "veggie", "mushroom": "veggie", "hamburger": "beef", "corn beef hash": "beef", "pulled pork": "pork"}


# def get_animal(meat):
#     return meat_to_animal_revised[meat]

temp_animal = df_food_revised["food"].map(lambda meat: meat_to_animal_revised[meat])

df_food_revised["animal"] =temp_animal 

df_food_revised
```

```python
# REPLACING VALUES

replaced_data = pd.Series([1., .999 , 2., .999, -1000, 3.])


print(replaced_data)

replaced_data.replace(.999, "not cool", inplace=True)

print(replaced_data)
```

```python
#REPLACING VALUES USING A DICTIONARY

replaced_data = pd.Series([1., .999 , 2., .999, -1000, 3.])

print(replaced_data)

replaced_data.replace({.999: "not cool", 3: "cool"}, inplace=True)

#print(replaced_data)    
```

```python
#DATA DICRETIZATIN AND BINNING

# ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
ages = [1, 21, 21, 51]
bins  = [18, 30, 40, 50, 80]




#THE RETURN VALUE HERE IS CALLED A CATEGORY! DIFFERENT THAN WHAT 
#WE HAVE USED UNTIL NOW
age_categories = pd.cut(ages, bins, precision=2)



age_categories.value_counts()
# pd.value_counts(age_categories)

# age_categories.codes


```

```python
#SPLIT THE DATA INTO QUARTERS

data = np.random.standard_normal((1000))

quartiles = pd.qcut(data, 4, precision=2)

quartiles.value_counts()
```

```python
#OUTLIERS

data_revised = pd.DataFrame(np.random.standard_normal((1000, 4)))

print(data_revised.describe())


col = data_revised[2]

print(col[col.abs() > 3])



```

```python
#MORE OUTLIERS FOR A DATAFRAME

data_revised[(data_revised.abs() > 3).any(axis="columns")].head()
```

```python
#PERMUTATION AND RANDOM SAMPLING

df_perm = pd.DataFrame(np.arange(5*7).reshape(5,7))

print(df_perm)

sampler = np.random.permutation(5)

print(sampler)


print(df_perm.take(sampler))
```

```python
#COOL IMPLEMENTATION USING DUMMIES AND CUT

values = np.random.uniform(size=10)

bins = [0, 0.2, 0.4, 0.6, 0.8, 1]

print(values)
print('\n')

print(pd.cut(values, bins))
print('\n')


pd.get_dummies(pd.cut(values, bins))
```
